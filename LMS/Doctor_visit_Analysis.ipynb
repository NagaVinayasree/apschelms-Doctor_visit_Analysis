# Plot horizontal barchart to analyse the reduced days of activity due to illness based on Gender

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'])
plt.title("Bar chart")
plt.xlabel("Reduced Activity")
plt.ylabel('Gender')
plt.show()
```

# Add age_category column 

```python
df['age_category'] = np.select([
    (df['age'] == 0.19),
    (df['age'] >=0.20) & (df['age'] <= 0.39),
    (df['age'] >=0.40) & (df['age'] <= 0.59),
    (df['age'] >= 0.60)
    ], ['Teen', 'Adult', 'Middle Adult', 'Senoir Adult'])
```

# illness based on age_category

```python
 df.groupby(["age_category",'illness']).size()
```

### No.of people are ill based on Age Category

```python
db=df.groupby('age_category')['illness'].sum().to_frame().reset_index()
plt.bar(db['age_category'],db['illness'])
plt.title("Bar chart")
plt.xlabel("Age Category")
plt.ylabel('illness')
plt.show()
```

# income based on age_category

```python
db=df.groupby('age_category')['income'].sum().to_frame().reset_index()
plt.plot(db['age_category'],db['income'])
plt.xlabel("Age Category")
plt.ylabel('income')
plt.show()
```
